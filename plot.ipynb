```python
import pandas as pd
import glob
import matplotlib.pyplot as plt
```

```python
path = r'ensembleResult/avg' # use your path
all_files = glob.glob(path + "/*.csv")

li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=1)
    li.append(df)

frame = pd.concat(li, axis=0, ignore_index=True)
frame =frame.rename(columns={"avg-fdl": "fdl .1", "avg-fdl.1": "fdl .25","avg-fdl.2": "fdl .5"})
frame
```

```python
frame[frame["Method"] == "function"]

fig = frame.boxplot(by='Method', 
                       column=["fdl .1","fdl .25","fdl .5"], 
                      figsize=(40,10),layout=(1,3))
```
